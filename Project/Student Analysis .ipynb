### Change the category value into int or float values 

```python
df2 = df.copy()
```

```python
#Changing all the binary-Category data into int 
Yes_no = {'yes': 1, 'no': 0}
df2['schoolsup'] = df2['schoolsup'].map(Yes_no)
df2['famsup'] = df2['famsup'].map(Yes_no)
df2['paid'] = df2['paid'].map(Yes_no)
df2['activities'] = df2['activities'].map(Yes_no)
df2['nursery'] = df2['nursery'].map(Yes_no)
df2['higher'] = df2['higher'].map(Yes_no)
df2['internet'] = df2['internet'].map(Yes_no)
df2['romantic'] = df2['romantic'].map(Yes_no)
```

```python
#map school data
school ={"GP": 1 , "MS": 0}
df2["school"] = df2["school"].map(school)


#map the sex data
sex = {'F': 1, 'M': 0}
df2['sex'] = df2['sex'].map(sex)

# map the address data
Add = {'U': 1, 'R': 0}
df2['address'] = df2['address'].map(Add)

# map the famili size data
Fam = {'LE3': 1, 'GT3': 0}
df2['famsize'] = df2['famsize'].map(Fam)

# map the parent's status
Pstat = {'T': 1, 'A': 0}
df2['Pstatus'] = df2['Pstatus'].map(Pstat)

# map the parent's job
Job = {'teacher': 0, 'health' : 1, 'services': 2,'at_home': 3,'other': 4}
df2['Mjob'] = df2['Mjob'].map(Job)
df2['Fjob'] = df2['Fjob'].map(Job)

# map the reason data
Reason = {'home': 0, 'reputation': 1, 'course': 2,'other': 3}
df2['reason'] = df2['reason'].map(Reason)

# map the guardian data
gdata = {'mother': 0, 'father': 1, 'other': 2}
df2['guardian'] = df2['guardian'].map(gdata)
```

###### Let us get grades to categorize students
Above 18 = Grade A

Between 14 and 18 = Grade B

Below 14 = Grade C

```python
def grades(df):
    Grade = []
    for i in df["GAvg"]:
        if (i >= 18):
            Grade.append("A")
        elif (i >= 14) and (i < 18):
            Grade.append("B")
        elif (i < 14):
            Grade.append("C")
    df["Grade"] = Grade
    return df           
```

```python
df3 = grades(df2)
df3.Grade
```

### Let us build a learning model 

```python
df4 = df3.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
Le_Grade = LabelEncoder()
df4["Grade"] = Le_Grade.fit_transform(df4["Grade"])
df4.head()
```

```python
X = df4.drop(["G1","G2","G3","GAvg","Grade"],axis="columns")
X.head()
```

```python
y = df4.Grade
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
DecisionTree = tree.DecisionTreeClassifier()
```

```python
DecisionTree.fit(X_train, y_train)
```

```python
DecisionTree.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

def model_sel(X, y):
    algos = {
      'decsion_tree' : {'model':DecisionTreeClassifier(), 'params': {}},
        'random_forest': {'model': RandomForestClassifier(), 'params' : {'n_estimators': [20,50,55]}},
        'lasso' : {'model':Lasso(), 'params':{'alpha': [1,2],'selection': ['random', 'cyclic']}}
    }
    
    scores = []

    cv = ShuffleSplit(n_splits=10 , test_size=0.2, random_state=10)

    for model_name, mp in algos.items():
        clf =  GridSearchCV(mp['model'], mp['params'], cv=cv, return_train_score=False)
        clf.fit(X, y)
        scores.append({
            'model': model_name,
            'best_score': clf.best_score_,
            'best_params': clf.best_params_
    })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
    
model_sel(X,y)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
scores1 = cross_val_score(RandomForestClassifier(n_estimators=50),X,y)
scores1
```

```python
np.average(scores1)
```
